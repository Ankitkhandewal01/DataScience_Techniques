```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset

titanic_data =pd.read_csv(r"C:\Users\Ankit\Desktop\All folders\BrainyBeam tasks\train.csv")

# Bar plot of passenger counts by sex
sex_counts = titanic_data['Sex'].value_counts()
plt.bar(sex_counts.index, sex_counts.values)
plt.xlabel('Sex')
plt.ylabel('Passenger Count')
plt.title('Passenger Counts by Sex')
plt.show()

# Pie chart of passenger survival
survival_counts = titanic_data['Survived'].value_counts()
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%')
plt.title('Passenger Survival')
plt.show()

# Histogram of passenger ages
plt.hist(titanic_data['Age'].dropna(), bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Passenger Age Distribution')
plt.show()

# Scatter plot of fare versus age
plt.scatter(titanic_data['Age'], titanic_data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs Age')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load Titanic dataset
data_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_data = pd.read_csv(data_url)

# Create a 3D scatter plot of age, fare, and passenger class
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(titanic_data['Age'], titanic_data['Fare'], titanic_data['Pclass'])

ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Pclass')
ax.set_title('Titanic Dataset - 3D Scatter Plot')

plt.show()
```
