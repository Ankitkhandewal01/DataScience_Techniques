# pivot(data,index,column,value)

```python
df['Age'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
pt=pd.pivot_table(df,index='Survived',columns='Sex',values='Age')
```

```python
pt
```

```python
sns.heatmap(pt,annot=True,cmap='coolwarm')
```
