# LogisticRegression

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

# Load the Iris dataset
X, y = load_iris(return_X_y=True)

# Create two logistic regression models with different values of C
model_regularized = LogisticRegression(C=0.1)
model_no_regularization = LogisticRegression(C=10)

# Fit the models to the data
model_regularized.fit(X, y)
model_no_regularization.fit(X, y)

# Print the coefficients
print("Regularized Model Coefficients:", model_regularized.coef_)
print("No Regularization Model Coefficients:", model_no_regularization.coef_)
```
